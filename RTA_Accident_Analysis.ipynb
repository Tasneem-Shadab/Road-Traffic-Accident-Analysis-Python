## phase 4: treatment of nulls

```python
obj_cols=RTA_df1.select_dtypes(include=['object']).columns
obj_cols
```

```python
obj_nan_cols=[col for col in obj_cols if RTA_df1[col].isnull().any()]
obj_nan_cols
```

```python
RTA_df1[obj_nan_cols].mode().value_counts().index
```

```python
RTA_df1[obj_nan_cols]=RTA_df1[obj_nan_cols].fillna(RTA_df1[obj_nan_cols].mode().iloc[0])
```

```python
int_cols=RTA_df1.select_dtypes(include=['int64']).columns
int_cols
```

```python
RTA_df1[int_cols]=RTA_df1[int_cols].fillna(RTA_df1[int_cols].median())
RTA_df1[int_cols]
```

## Phase 6: Feature enginreering

```python
RTA_df1['Hour'] = RTA_df1['Time'].dt.hour
RTA_df1["Hour"]
```

```python
RTA_df1['Time'].head(10)
```

```python
RTA_df1["Month"]=RTA_df1['Time'].dt.month_name()
```

```python
RTA_df1["Month"]
```

```python
RTA_df1["day"]=RTA_df1['Time'].dt.day_name()
RTA_df1["day"]
```

```python
def get_time_period(hour):
    if 0 <= hour < 6:
        return 'Night'
    elif 6 <= hour < 12:
        return 'Morning'
    elif 12 <= hour < 18:
        return 'Afternoon'
    else:
        return 'Evening'

RTA_df1['Time_Period'] = RTA_df1['Hour'].apply(get_time_period)
```

## Phase 9: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## 1. Bar Chart: Accidents by Hour

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=RTA_df1,
    x='Hour',
    color='steelblue'
)

plt.title('Accidents by Hour')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=0)
plt.show()
```

## 2. Count Plot: Vehicle Type

```python
plt.figure(figsize=(12,6))

sns.countplot(
    data=RTA_df1,
    y='Type_of_vehicle',
    order=RTA_df1['Type_of_vehicle'].value_counts().index,
    palette='viridis'
)

plt.title('Accidents by Vehicle Type')
plt.xlabel('Number of Accidents')
plt.ylabel('Vehicle Type')

plt.show()
```

## 3. Heatmap: Accident Severity vs Weather Conditions

```python
severity_weather = pd.crosstab(
    RTA_df1['Weather_conditions'],
    RTA_df1['Accident_severity']
)

plt.figure(figsize=(10,6))

sns.heatmap(
    severity_weather,
    annot=True,
    cmap='YlOrRd',
    fmt='d'
)

plt.title('Accident Severity vs Weather Conditions')
plt.xlabel('Accident Severity')
plt.ylabel('Weather Conditions')

plt.show()
```

 ## 4. Distribution Plot: Number of Casualties

```python
plt.figure(figsize=(10,6))

sns.histplot(
    RTA_df1['Number_of_casualties'],
    bins=10,
    kde=True,
    color='purple'
)

plt.title('Distribution of Number of Casualties')
plt.xlabel('Number of Casualties')
plt.ylabel('Frequency')

plt.show()
```
